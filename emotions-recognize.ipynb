```python
import cv2
from ultralytics import YOLO
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
import random
from glob import glob
import warnings
warnings.simplefilter("ignore")
```

```python
model = YOLO("yolov8n-cls.pt")
```

```python
emotions_folder ="/kaggle/input/emotions/Emotions/train"
emotions = ["sad", "happy", "angry", "disgust", "neutral", "fear", "surprise"]
plt.figure(figsize=(15, 15))
img_idx = 1
for emotion in emotions:
    emotion_path = os.path.join(emotions_folder, emotion)
    image_files = os.listdir(emotion_path)
    selected_images = random.sample(image_files, 2)    
    for img_file in selected_images:
        img_path = os.path.join(emotion_path, img_file)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        plt.subplot(len(emotions), 2, img_idx)
        plt.imshow(img)
        plt.title(f"{emotion.capitalize()}")
        plt.axis('off')       
        img_idx += 1
plt.tight_layout()
plt.show()
```

```python
Dataset = "/kaggle/input/emotions/Emotions"
results = model.train(data=Dataset, epochs=15, imgsz=64, batch=-1)
```
